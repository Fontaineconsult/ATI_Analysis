```python
from neomodelschema import set_connection, Goal, ATISubCommittee
set_connection()

import datetime

def add_goal(name,
             removed,
             goal,
             goal_number,
             part_of_subcommittee,
             ):
    

    new_goal = Goal(
        name=name,
        removed=removed,
        goal=goal,
        goal_number=goal_number,
        date_added=datetime.datetime.utcnow()
    )

    new_goal.save()
    sub_committee = ATISubCommittee.get_or_create({'name': part_of_subcommittee})
    print(sub_committee)
    new_goal.part_of_atisubcommittee.connect(sub_committee[0])
    sub_committee[0].goals.connect(new_goal)


add_goal(
    name="ATI Instructional Materials Accessibility Plan",
    removed=False,
    goal="Annually review and update the ATI Instructional Materials Accessibility Plan.",
    goal_number=9,
    part_of_subcommittee="Instructional Materials",

)
```

```python
from neomodelschema import set_connection, SuccessIndicator, Goal
from decimal import Decimal
set_connection()

import datetime

def add_success_indicator(number,
                          removed,
                          associated_goal,
                          success_indicator,
                          sub_committee,
             ):


    new_indicator = SuccessIndicator(
        number=number,
        removed=removed,
        success_indicator=success_indicator,
        date_added=datetime.datetime.utcnow(),
        composite_key = f"{number}-{sub_committee}"
    )

    
    try:
        goal = Goal.nodes.get(name=associated_goal)
    except:
        print("Goal Doesn't Exist")
        return 
    new_indicator.save()
    print(goal)
    new_indicator.associated_goal.connect(goal)
    goal.success_indicators.connect(new_indicator)


add_success_indicator(
    number=Decimal('9.2'),
    removed=False,
    associated_goal="ATI Procurement Plan",
    success_indicator="Developed a process for the procurement subcommittee to inform the revision and coordinate the implementation of Annual Procurement Plan.",
    sub_committee="Procurement"
    

)
```

```python
from neomodelschema import set_connection, YearSuccessEvidence, StatusLevel, AcademicYear,SuccessIndicator
set_connection()

def add_year_success_evidence(academic_year, status_level, related_success_indicator):
    
    new_year_status = YearSuccessEvidence(

        year_identifier=f"{academic_year}-{related_success_indicator}",
       
        
    )
    

    try:
        AY = AcademicYear.nodes.get(name=academic_year)
        print(AY)
        status_level = StatusLevel.nodes.get(status_level=status_level)
        print(status_level)
        success_indicator = SuccessIndicator.nodes.get(composite_key=related_success_indicator)
        print(success_indicator)

    except:
        print("missing connecting node")
        return
    new_year_status.save()
    new_year_status.academic_year.connect(AY),
    new_year_status.status_level.connect(status_level),
    new_year_status.related_success_indicator.connect(success_indicator)

add_year_success_evidence("2020/2021", "Established", "9.2-Procurement")
    
```
