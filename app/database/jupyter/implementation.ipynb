```python
import sys
sys.path.append(r"C:\Users\Fonta\IdeaProjects\ATI_Analysis")
sys.path.append(r"C:\Users\913678186\IdeaProjects\ATI_Analysis")
from app.database.queries.governance.create import add_guideline
###
### ADD GUIDELINE
###

import ipywidgets as widgets
from IPython.display import display
from datetime import datetime
from app.database.queries.implementation.create import add_guidance


# Create form widgets for adding a Guideline
guideline_title = widgets.Text(value='', placeholder='Enter guideline title', description='Title:', disabled=False)
guideline_description = widgets.Textarea(value='', placeholder='Enter guideline description', description='Description:', disabled=False)
guideline_effective_date = widgets.Text(value='', placeholder='YYYY-MM-DD', description='Effective Date:', disabled=False)
guideline_last_updated = widgets.Text(value='', placeholder='YYYY-MM-DD', description='Last Updated:', disabled=False)
submit_button = widgets.Button(description='Submit', disabled=False, button_style='', tooltip='Click to submit', icon='check')
output = widgets.Output()

# Define the event handler for the submit button
def on_button_click(b):
    with output:
        output.clear_output()
        success = add_guideline(
            title=guideline_title.value,
            description=guideline_description.value,
            effective_date=datetime.strptime(guideline_effective_date.value, "%Y-%m-%d").date() if guideline_effective_date.value else None,
            last_updated=datetime.strptime(guideline_last_updated.value, "%Y-%m-%d").date() if guideline_last_updated.value else None,
        )
        if success:
            print('Guideline added successfully')
        else:
            print('Failed to add guideline')

# Attach the event handler to the button
submit_button.on_click(on_button_click)

# Display the form
form_items = [guideline_title, guideline_description, guideline_effective_date, guideline_last_updated, submit_button, output]
form = widgets.VBox(form_items)
display(form)
```
