## CREATE Operations

### Add a Document

Parameters:
- `name` (str, required)
- `file_path` (str, optional) - local file path
- `uri_path` (str, optional) - URI/URL (at least one of file_path or uri_path required)
- `depreciated` (bool, default False)
- `depreciated_date` (str, YYYY-MM-DD, optional)
- `is_administrative_review_documentation` (bool, default False)
- `is_milestone_and_measures_documentation` (bool, default False)
- `implementation_id` (str, optional) - unique_id of implementation to assign to
- `implementation_type` (str, optional) - e.g. "Process", "Project"
- `academic_year` (str, optional) - e.g. "2024-2025"
- `include_in_year` (bool, default True)

```python
from app.database.queries.documentation.create import add_document

add_document(
    name='Example Document',
    uri_path='https://example.com/doc.pdf',
    depreciated=False,
    is_administrative_review_documentation=False,
    is_milestone_and_measures_documentation=False
)
```

### Add a Webpage

Parameters:
- `url` (str, required, unique)
- `name` (str, required)
- `description` (str)
- `no_longer_exists` (bool)
- `depreciated` (bool)
- `depreciated_date` (str, YYYY-MM-DD)
- `include_in_report` (bool)
- `created_by` (str, optional) - employee_id or unique_id of Person
- `implementation_id`, `implementation_type`, `academic_year`, `include_in_year` (optional)

```python
from app.database.queries.documentation.create import add_webpage

add_webpage(
    url='https://example.com/accessibility',
    name='Example Webpage',
    description='An accessibility resource page',
    no_longer_exists=False,
    depreciated=False,
    depreciated_date=None,
    include_in_report=True
)
```
