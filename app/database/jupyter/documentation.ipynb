```python
import ipywidgets as widgets
from IPython.display import display
from app.database.queries.documentation.create import add_document

# Create form widgets
document_name = widgets.Text(value='', placeholder='Enter document name', description='Name:', disabled=False)
document_file_path = widgets.Text(value='', placeholder='Enter file path', description='File Path:', disabled=False)
document_uri_path = widgets.Text(value='', placeholder='Enter URI path', description='URI Path:', disabled=False)
document_is_admin_review = widgets.Checkbox(value=False, description='Administrative Review Documentation', disabled=False)
document_is_milestone_measures = widgets.Checkbox(value=False, description='Milestone and Measures Documentation', disabled=False)
submit_button = widgets.Button(description='Submit', disabled=False, button_style='', tooltip='Click to submit', icon='check')
output = widgets.Output()

# Define the event handler for the submit button
def on_button_click(b):
    with output:
        output.clear_output()
        success = add_document(
            name=document_name.value,
            file_path=document_file_path.value,
            uri_path=document_uri_path.value,
            is_administrative_review_documentation=document_is_admin_review.value,
            is_milestone_and_measures_documentation=document_is_milestone_measures.value
        )
        if success:
            print('Document added successfully')
        else:
            print('Failed to add document')

# Attach the event handler to the button
submit_button.on_click(on_button_click)

# Display the form
form_items = [document_name, document_file_path, document_uri_path, document_is_admin_review, document_is_milestone_measures, submit_button, output]
form = widgets.VBox(form_items)
display(form)
```
