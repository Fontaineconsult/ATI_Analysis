```python
#
# Assign Documentation to Implementation
#
import sys
sys.path.append(r"C:\Users\Fonta\IdeaProjects\ATI_Analysis")
sys.path.append(r"C:\Users\913678186\IdeaProjects\ATI_Analysis")

import ipywidgets as widgets
from IPython.display import display
from app.database.queries.documentation.read import (
    get_all_documents,
    get_all_webpages,
    get_all_notes,
    get_all_messages,
    get_all_metrics
)
from app.database.class_factory import documentation_types, implementation_types
from app.database.queries.tools.utilities import get_all_implementations
from app.database.queries.implementation.update import assign_documentation_to_implementation

# Adjusted widget widths
widget_width = '400px'
button_width = '200px'

# Retrieve documentation data
documentation = {
    'document': get_all_documents(),
    'webpage': get_all_webpages(),
    'note': get_all_notes(),
    'message': get_all_messages(),
    'metric': get_all_metrics()
}

# Create a dropdown menu for documentation types
doc_type_select = widgets.Dropdown(
    options=documentation_types,
    value=documentation_types[0],  # default value
    description='Doc Type:',
    layout=widgets.Layout(width=widget_width)
)

# Initialize doc_select options based on the initial doc_type_select value
initial_docs = documentation[doc_type_select.value]
initial_options = [doc.name for doc in initial_docs]  # Adjust attribute as needed
doc_select = widgets.Dropdown(
    options=initial_options,
    value=initial_options[0] if initial_options else None,
    description='Document:',
    layout=widgets.Layout(width=widget_width)
)

# Create the dropdown for implementation types
implementation_dropdown = widgets.Dropdown(
    options=implementation_types,  # Add 'plan' if needed
    value='process',
    description='Impl Type:',
    layout=widgets.Layout(width=widget_width)
)

# Create the dropdown for implementation outputs
implementation_output_dropdown = widgets.Dropdown(
    options=[],
    description='Implementation:',
    layout=widgets.Layout(width=widget_width)
)

def populate_implementation_output_dropdown(implementation_type):
    # Fetch implementations based on the selected type
    implementations = get_all_implementations(implementation_type)
    # Update the options of implementation_output_dropdown
    implementation_output_dropdown.options = [(imp.title, imp.title) for imp in implementations]
    implementation_output_dropdown.value = implementations[0].title if implementations else None

# Define a function that will be triggered when the implementation_dropdown value changes
def on_implementation_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        implementation_type = change['new']
        populate_implementation_output_dropdown(implementation_type)

# Define the event handler function for documentation type changes
def on_doc_type_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        new_type = change['new']
        # Get the new list of documents for the selected type
        new_docs = documentation[new_type]
        # Extract titles or names for the dropdown options
        new_options = [doc.name for doc in new_docs]  # Adjust attribute as needed
        # Update doc_select options and value
        doc_select.options = new_options
        doc_select.value = new_options[0] if new_options else None

# Attach the event handlers
doc_type_select.observe(on_doc_type_change, names='value')
implementation_dropdown.observe(on_implementation_change, names='value')

# Populate the implementation output dropdown initially
populate_implementation_output_dropdown(implementation_dropdown.value)

# Create the Assign Documentation button
assign_documentation_button = widgets.Button(
    description='Assign Documentation',
    button_style='success',
    layout=widgets.Layout(width=button_width)
)

# Define the event handler for the button
def on_assign_documentation_button_clicked(button):
    # Retrieve values from dropdown menus
    implementation_title = implementation_output_dropdown.value
    implementation_type = implementation_dropdown.value
    documentation_type = doc_type_select.value
    documentation_title = doc_select.value

    # Call the function with the retrieved values

    assign_documentation_to_implementation(
        implementation_title=implementation_title,
        implementation_type=implementation_type,
        documentation_type=documentation_type,
        documentation_title=documentation_title
    )
    print(f"Successfully assigned '{documentation_title}' to '{implementation_title}'")


# Attach the event handler to the button
assign_documentation_button.on_click(on_assign_documentation_button_clicked)

# Arrange all widgets in a single vertical layout
vertical_layout = widgets.VBox([
    doc_type_select,
    doc_select,
    implementation_dropdown,
    implementation_output_dropdown,
    assign_documentation_button
])

# Display the vertical layout
display(vertical_layout)



```
