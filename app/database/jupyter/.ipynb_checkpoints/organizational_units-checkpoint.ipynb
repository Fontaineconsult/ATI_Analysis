### List Employees in a Department

```python
from app.database.graph_schema import Department

dept_name = 'PASTE_DEPARTMENT_NAME_HERE'  # <-- change this

try:
    dept = Department.nodes.get(name=dept_name)
    employees = dept.employs.all()
    print(f'=== {dept.name} ({len(employees)} employees) ===')
    if employees:
        df = pd.DataFrame([{
            'name': e.name,
            'employee_id': e.employee_id,
            'title': e.title,
            'email': e.email,
            'active': e.active
        } for e in employees])
        display(df)
    else:
        print('  No employees assigned.')
except Exception as e:
    print(f'Error: {e}')
```

### Organizational Units Summary

Shows employee count for each department, college, and vendor.

```python
from app.database.queries.organizational_units.read import get_all_departments, get_all_colleges, get_all_vendors

print('=== Departments ===')
for dept in get_all_departments():
    emp_count = len(dept.employs.all())
    print(f'  {dept.name} ({dept.location}) - {emp_count} employees')

print('\n=== Colleges ===')
for college in get_all_colleges():
    emp_count = len(college.employs.all())
    print(f'  {college.name} ({college.location}) - {emp_count} employees')

print('\n=== Vendors ===')
for vendor in get_all_vendors():
    emp_count = len(vendor.employs.all())
    print(f'  {vendor.name} ({vendor.location}) - {emp_count} employees')
```

## RELATIONSHIP Management

### View All Relationships for a Department

```python
from app.database.graph_schema import Department

dept = Department.nodes.first()  # <-- change lookup as needed
if dept:
    print(f'=== {dept.name} ===')
    print(f'  Location: {dept.location}')
    print(f'  unique_id: {dept.unique_id}')
    
    employees = dept.employs.all()
    print(f'\nEmployees ({len(employees)}):')
    for e in employees:
        print(f'  - {e.name} ({e.employee_id})')
    
    yses = dept.implements_yse.all()
    print(f'\nImplements YSE ({len(yses)}):')
    for yse in yses:
        print(f'  - {yse.year_identifier}')
    
    campuses = dept.operates_under_campus.all()
    print(f'\nCampus ({len(campuses)}):')
    for c in campuses:
        print(f'  - {c.name}')
else:
    print('No Department nodes found.')
```

```python
from app.database.graph_schema import College

college = College.nodes.first()  # <-- change lookup as needed
if college:
    print(f'=== {college.name} ===')
    print(f'  Location: {college.location}')
    print(f'  unique_id: {college.unique_id}')
    
    employees = college.employs.all()
    print(f'\nEmployees ({len(employees)}):')
    for e in employees:
        print(f'  - {e.name} ({e.employee_id})')
    
    yses = college.implements_yse.all()
    print(f'\nImplements YSE ({len(yses)}):')
    for yse in yses:
        print(f'  - {yse.year_identifier}')
    
    campuses = college.operates_under_campus.all()
    print(f'\nCampus ({len(campuses)}):')
    for c in campuses:
        print(f'  - {c.name}')
else:
    print('No College nodes found.')
```
