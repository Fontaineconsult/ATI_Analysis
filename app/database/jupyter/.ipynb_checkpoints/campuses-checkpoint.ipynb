# Campuses Admin Notebook

Manage **Campus** nodes and their relationships to Departments and Colleges.

A Campus represents a physical location within the university system. Departments and Colleges operate under a Campus via the `operates_under_campus` relationship.

**Campus properties:**
- `unique_id` (auto-generated)
- `name` (str, unique)

### List Departments Under a Campus

```python
from app.database.queries.organizational_units.read import get_departments_by_campus

campus_name = 'PASTE_CAMPUS_NAME_HERE'  # <-- change this

departments = get_departments_by_campus(campus_name)
print(f'=== Departments under {campus_name} ({len(departments)}) ===')
if departments:
    df = pd.DataFrame([{
        'unique_id': d.unique_id,
        'name': d.name,
        'location': d.location,
        'employees': len(d.employs.all())
    } for d in departments])
    display(df)
else:
    print('  No departments connected to this campus.')
```

### List Colleges Under a Campus

```python
from app.database.queries.organizational_units.read import get_colleges_by_campus

campus_name = 'PASTE_CAMPUS_NAME_HERE'  # <-- change this

colleges = get_colleges_by_campus(campus_name)
print(f'=== Colleges under {campus_name} ({len(colleges)}) ===')
if colleges:
    df = pd.DataFrame([{
        'unique_id': c.unique_id,
        'name': c.name,
        'location': c.location,
        'employees': len(c.employs.all())
    } for c in colleges])
    display(df)
else:
    print('  No colleges connected to this campus.')
```

### Campus Summary

Shows department and college counts for each campus, plus unassigned units.

```python
from app.database.queries.organizational_units.read import (
    get_all_campuses, get_all_departments, get_all_colleges
)

campuses = get_all_campuses()
all_departments = get_all_departments()
all_colleges = get_all_colleges()

if not campuses:
    print('No Campus nodes found.')
else:
    for campus in campuses:
        depts = [d for d in all_departments if d.operates_under_campus.is_connected(campus)]
        cols = [c for c in all_colleges if c.operates_under_campus.is_connected(campus)]
        total_employees = sum(len(d.employs.all()) for d in depts) + sum(len(c.employs.all()) for c in cols)
        print(f'=== {campus.name} ===')
        print(f'  Departments: {len(depts)}')
        print(f'  Colleges: {len(cols)}')
        print(f'  Total employees (via depts + colleges): {total_employees}')
        print()

# Show unassigned departments/colleges
unassigned_depts = [d for d in all_departments if len(d.operates_under_campus.all()) == 0]
unassigned_cols = [c for c in all_colleges if len(c.operates_under_campus.all()) == 0]
if unassigned_depts or unassigned_cols:
    print('=== Not Assigned to Any Campus ===')
    for d in unassigned_depts:
        print(f'  [Dept] {d.name} ({d.location})')
    for c in unassigned_cols:
        print(f'  [College] {c.name} ({c.location})')
```
