## RELATIONSHIP Management

All governance nodes share the same relationship pattern:
- `informed_goals` -> Goal ("informs")
- `supporting_documents` -> Document ("is_documented_by")
- `supporting_websites` -> Webpage ("is_documented_by")
- `supporting_notes` -> Note ("is_documented_by")
- `supporting_messages` -> Message ("is_documented_by")

### View Relationships for a Governance Node

```python
from app.database.graph_schema import Law

node = Law.nodes.first()  # <-- change class and lookup as needed
if node:
    print(f'=== {node.title} ===')
    print(f'\nInformed Goals ({len(node.informed_goals.all())}):')
    for g in node.informed_goals.all():
        print(f'  - Goal #{g.goal_number}: {g.name}')
    print(f'\nSupporting Documents ({len(node.supporting_documents.all())}):')
    for d in node.supporting_documents.all():
        print(f'  - {d.name}')
    print(f'\nSupporting Websites ({len(node.supporting_websites.all())}):')
    for w in node.supporting_websites.all():
        print(f'  - {w.name} ({w.url})')
    print(f'\nSupporting Notes ({len(node.supporting_notes.all())}):')
    for n in node.supporting_notes.all():
        print(f'  - {n.name}')
    print(f'\nSupporting Messages ({len(node.supporting_messages.all())}):')
    for m in node.supporting_messages.all():
        print(f'  - {m.name}')
else:
    print('No nodes found.')
```

### Connect Governance Node to a Goal

```python
from app.database.graph_schema import Law, Goal

law = Law.nodes.get(title='PASTE_TITLE_HERE')  # <-- change
goal = Goal.nodes.get(goal_number=1)  # <-- change

if not law.informed_goals.is_connected(goal):
    law.informed_goals.connect(goal)
    print(f'Connected "{law.title}" -> Goal #{goal.goal_number}')
else:
    print('Already connected.')
```
