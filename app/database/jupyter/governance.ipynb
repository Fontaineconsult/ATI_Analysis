```python
"""
ADD LAW
"""

# Import necessary modules
import ipywidgets as widgets
from IPython.display import display
from app.database.queries.governance.create import add_law
from datetime import datetime

# Create form widgets for adding a Law
law_title = widgets.Text(value='', placeholder='Enter law title', description='Title:', disabled=False)
law_description = widgets.Textarea(value='', placeholder='Enter law description', description='Description:', disabled=False)
law_effective_date = widgets.Text(value='', placeholder='YYYY-MM-DD', description='Effective Date:', disabled=False)
law_last_updated = widgets.Text(value='', placeholder='YYYY-MM-DD', description='Last Updated:', disabled=False)
law_relevant_sections = widgets.Text(value='', placeholder='Enter relevant sections', description='Sections:', disabled=False)
law_legislative_authority = widgets.Text(value='', placeholder='Enter legislative authority', description='Authority:', disabled=False)
submit_button = widgets.Button(description='Submit', disabled=False, button_style='', tooltip='Click to submit', icon='check')
output = widgets.Output()

# Define the event handler for the submit button
def on_button_click(b):
    with output:
        output.clear_output()
        success = add_law(
            title=law_title.value,
            description=law_description.value,
            effective_date=datetime.strptime(law_effective_date.value, "%Y-%m-%d").date(),
            last_updated=datetime.strptime(law_last_updated.value, "%Y-%m-%d").date(),
            relevant_sections=law_relevant_sections.value,
            legislative_authority=law_legislative_authority.value
        )
        if success:
            print('Law added successfully')
        else:
            print('Failed to add law')

# Attach the event handler to the button
submit_button.on_click(on_button_click)

# Display the form
form_items = [law_title, law_description, law_effective_date, law_last_updated, law_relevant_sections, law_legislative_authority, submit_button, output]
form = widgets.VBox(form_items)
display(form)

""" 
ADD SUPPORTING DOCUMENT
"""



```
