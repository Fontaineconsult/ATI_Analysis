### Organizational Units Summary

Shows employee count for each department, college, and vendor, plus campus assignments.

```python
from app.database.queries.organizational_units.read import (
    get_all_departments, get_all_colleges, get_all_vendors, get_all_campuses
)

print('=== Departments ===')
for dept in get_all_departments():
    emp_count = len(dept.employs.all())
    print(f'  {dept.name} ({dept.location}) - {emp_count} employees')

print('\n=== Colleges ===')
for college in get_all_colleges():
    emp_count = len(college.employs.all())
    print(f'  {college.name} ({college.location}) - {emp_count} employees')

print('\n=== Vendors ===')
for vendor in get_all_vendors():
    emp_count = len(vendor.employs.all())
    print(f'  {vendor.name} ({vendor.location}) - {emp_count} employees')

print('\n=== Campuses ===')
all_depts = get_all_departments()
all_cols = get_all_colleges()
for campus in get_all_campuses():
    depts = [d for d in all_depts if d.operates_under_campus.is_connected(campus)]
    cols = [c for c in all_cols if c.operates_under_campus.is_connected(campus)]
    print(f'  {campus.name} - {len(depts)} departments, {len(cols)} colleges')

unassigned_depts = [d for d in all_depts if len(d.operates_under_campus.all()) == 0]
unassigned_cols = [c for c in all_cols if len(c.operates_under_campus.all()) == 0]
if unassigned_depts or unassigned_cols:
    print('\n=== Not Assigned to Any Campus ===')
    for d in unassigned_depts:
        print(f'  [Dept] {d.name} ({d.location})')
    for c in unassigned_cols:
        print(f'  [College] {c.name} ({c.location})')
```

### Assign an Employee to a Department

Select a department and employee from the dropdowns, then click **Assign**.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from app.database.queries.organizational_units.read import get_all_departments
from app.database.queries.organizational_units.update import assign_employee_to_department
from app.database.graph_schema import Person

# Build dropdown options
dept_options = {d.name: d.name for d in get_all_departments()}
person_list = Person.nodes.all()
employee_options = {f"{p.name} ({p.employee_id})": p.employee_id for p in person_list}

dept_dropdown = widgets.Dropdown(
    options=dept_options,
    description='Department:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

employee_dropdown = widgets.Dropdown(
    options=employee_options,
    description='Employee:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

assign_btn = widgets.Button(
    description='Assign',
    button_style='success',
    icon='check'
)

output = widgets.Output()

def on_assign_click(b):
    with output:
        clear_output()
        try:
            assign_employee_to_department(
                department_name=dept_dropdown.value,
                employee_id=employee_dropdown.value
            )
        except Exception as e:
            print(f'Error: {e}')

assign_btn.on_click(on_assign_click)

display(dept_dropdown, employee_dropdown, assign_btn, output)
```

## RELATIONSHIP Management

### View All Relationships for a Department

```python
from app.database.graph_schema import Department

dept = Department.nodes.first()  # <-- change lookup as needed
if dept:
    print(f'=== {dept.name} ===')
    print(f'  Location: {dept.location}')
    print(f'  unique_id: {dept.unique_id}')
    
    employees = dept.employs.all()
    print(f'\nEmployees ({len(employees)}):')
    for e in employees:
        print(f'  - {e.name} ({e.employee_id})')
    
    yses = dept.implements_yse.all()
    print(f'\nImplements YSE ({len(yses)}):')
    for yse in yses:
        print(f'  - {yse.year_identifier}')
    
    campuses = dept.operates_under_campus.all()
    print(f'\nCampus ({len(campuses)}):')
    for c in campuses:
        print(f'  - {c.name}')
else:
    print('No Department nodes found.')
```
