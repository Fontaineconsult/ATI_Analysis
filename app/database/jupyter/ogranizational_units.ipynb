```python
###
### Add and Delete Organizational Unit
###

# Import necessary modules
import ipywidgets as widgets
from IPython.display import display
from app.database.queries.organizational_units.create import add_organizational_unit
from app.database.queries.delete import delete_organizational_unit
from neomodel import config

# Configure the database
config.DATABASE_URL = "bolt://neo4j:accessibility@localhost:7687"

# Create form widgets for adding an organizational unit
unit_name = widgets.Text(value='', placeholder='Enter unit name', description='Name:', disabled=False)
unit_location = widgets.Text(value='', placeholder='Enter unit location', description='Location:', disabled=False)
submit_button = widgets.Button(description='Submit', disabled=False, button_style='', tooltip='Click to submit', icon='check')
output = widgets.Output()

# Define the event handler for the submit button
def on_button_click(b):
    with output:
        output.clear_output()
        success = add_organizational_unit(name=unit_name.value, location=unit_location.value)
        if success:
            print('Organizational unit added successfully')
        else:
            print('Failed to add organizational unit')

# Attach the event handler to the button
submit_button.on_click(on_button_click)

# Create form widgets for deleting an organizational unit
delete_name = widgets.Text(value='', placeholder='Enter unit name', description='Name:', disabled=False)
delete_button = widgets.Button(description='Delete', disabled=False, button_style='', tooltip='Click to delete', icon='trash')

# Define the event handler for the delete button
def on_delete_button_click(b):
    with output:
        output.clear_output()
        success = delete_organizational_unit(name=delete_name.value)
        if success:
            print('Organizational unit deleted successfully')
        else:
            print('Failed to delete organizational unit')

# Attach the event handler to the delete button
delete_button.on_click(on_delete_button_click)

# Display the forms
add_form_items = [unit_name, unit_location, submit_button, output]
delete_form_items = [delete_name, delete_button, output]

add_form = widgets.VBox(add_form_items)
delete_form = widgets.VBox(delete_form_items)

display(add_form)
display(delete_form)
```
