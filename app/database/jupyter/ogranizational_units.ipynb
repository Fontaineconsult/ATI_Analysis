```python
###
### Add and Delete Organizational Unit
###

# Import necessary modules
import ipywidgets as widgets
from IPython.display import display
from app.database.queries.organizational_units.create import add_department, add_college, add_vendor
from app.database.queries.organizational_units.delete import delete_department, delete_college, delete_vendor
from neomodel import config

# Configure the database
config.DATABASE_URL = "bolt://neo4j:accessibility@localhost:7687"

# Create form widgets for adding a department, college, and vendor
department_name = widgets.Text(value='', placeholder='Enter department name', description='Name:', disabled=False)
department_location = widgets.Text(value='', placeholder='Enter department location', description='Location:', disabled=False)
submit_department_button = widgets.Button(description='Submit', disabled=False, button_style='', tooltip='Click to submit', icon='check')

college_name = widgets.Text(value='', placeholder='Enter college name', description='Name:', disabled=False)
college_location = widgets.Text(value='', placeholder='Enter college location', description='Location:', disabled=False)
submit_college_button = widgets.Button(description='Submit', disabled=False, button_style='', tooltip='Click to submit', icon='check')

vendor_name = widgets.Text(value='', placeholder='Enter vendor name', description='Name:', disabled=False)
vendor_location = widgets.Text(value='', placeholder='Enter vendor location', description='Location:', disabled=False)
submit_vendor_button = widgets.Button(description='Submit', disabled=False, button_style='', tooltip='Click to submit', icon='check')

output = widgets.Output()

# Define the event handler for the submit button
def on_department_button_click(b):
    with output:
        output.clear_output()
        success = add_department(name=department_name.value, location=department_location.value)
        if success:
            print('Department added successfully')
        else:
            print('Failed to add department')

def on_college_button_click(b):
    with output:
        output.clear_output()
        success = add_college(name=college_name.value, location=college_location.value)
        if success:
            print('College added successfully')
        else:
            print('Failed to add college')

def on_vendor_button_click(b):
    with output:
        output.clear_output()
        success = add_vendor(name=vendor_name.value, location=vendor_location.value)
        if success:
            print('Vendor added successfully')
        else:
            print('Failed to add vendor')

# Attach the event handler to the button
submit_department_button.on_click(on_department_button_click)
submit_college_button.on_click(on_college_button_click)
submit_vendor_button.on_click(on_vendor_button_click)

# Create form widgets for deleting a department, college, and vendor
delete_department_name = widgets.Text(value='', placeholder='Enter department name', description='Name:', disabled=False)
delete_department_button = widgets.Button(description='Delete', disabled=False, button_style='', tooltip='Click to delete', icon='trash')

delete_college_name = widgets.Text(value='', placeholder='Enter college name', description='Name:', disabled=False)
delete_college_button = widgets.Button(description='Delete', disabled=False, button_style='', tooltip='Click to delete', icon='trash')

delete_vendor_name = widgets.Text(value='', placeholder='Enter vendor name', description='Name:', disabled=False)
delete_vendor_button = widgets.Button(description='Delete', disabled=False, button_style='', tooltip='Click to delete', icon='trash')

# Define the event handler for the delete button
def on_delete_department_button_click(b):
    with output:
        output.clear_output()
        success = delete_department(name=delete_department_name.value)
        if success:
            print('Department deleted successfully')
        else:
            print('Failed to delete department')

def on_delete_college_button_click(b):
    with output:
        output.clear_output()
        success = delete_college(name=delete_college_name.value)
        if success:
            print('College deleted successfully')
        else:
            print('Failed to delete college')

def on_delete_vendor_button_click(b):
    with output:
        output.clear_output()
        success = delete_vendor(name=delete_vendor_name.value)
        if success:
            print('Vendor deleted successfully')
        else:
            print('Failed to delete vendor')

# Attach the event handler to the delete button
delete_department_button.on_click(on_delete_department_button_click)
delete_college_button.on_click(on_delete_college_button_click)
delete_vendor_button.on_click(on_delete_vendor_button_click)

# Display the forms
add_department_form_items = [department_name, department_location, submit_department_button, output]
delete_department_form_items = [delete_department_name, delete_department_button, output]

add_college_form_items = [college_name, college_location, submit_college_button, output]
delete_college_form_items = [delete_college_name, delete_college_button, output]

add_vendor_form_items = [vendor_name, vendor_location, submit_vendor_button, output]
delete_vendor_form_items = [delete_vendor_name, delete_vendor_button, output]

add_department_form = widgets.VBox(add_department_form_items)
delete_department_form = widgets.VBox(delete_department_form_items)

add_college_form = widgets.VBox(add_college_form_items)
delete_college_form = widgets.VBox(delete_college_form_items)

add_vendor_form = widgets.VBox(add_vendor_form_items)
delete_vendor_form = widgets.VBox(delete_vendor_form_items)

display(add_department_form)
display(delete_department_form)

display(add_college_form)
display(delete_college_form)

display(add_vendor_form)
display(delete_vendor_form)
```

```python
import ipywidgets as widgets
from IPython.display import display
from app.database.queries.individuals.read import get_all_individuals
from app.database.queries.organizational_units.read import get_all_departments, get_all_colleges, get_all_vendors
from app.database.queries.organizational_units.update import assign_employee_to_department, assign_employee_to_college, assign_employee_to_vendor

# Retrieve the list of all individuals, departments, colleges, and vendors
individuals = get_all_individuals()
departments = get_all_departments()
colleges = get_all_colleges()
vendors = get_all_vendors()

# Create dropdown widgets for individuals, departments, colleges, and vendors
individual_dropdown = widgets.Dropdown(
    options=[(individual.name, individual.employee_id) for individual in individuals],
    description='Individual:',
)

department_dropdown = widgets.Dropdown(
    options=[department.name for department in departments],
    description='Department:',
)

college_dropdown = widgets.Dropdown(
    options=[college.name for college in colleges],
    description='College:',
)

vendor_dropdown = widgets.Dropdown(
    options=[vendor.name for vendor in vendors],
    description='Vendor:',
)

# Create buttons for the assignment actions
assign_to_department_button = widgets.Button(description="Assign to Department")
assign_to_college_button = widgets.Button(description="Assign to College")
assign_to_vendor_button = widgets.Button(description="Assign to Vendor")

# Event handler for the department assignment button click
def on_assign_to_department_clicked(b):
    employee_id = individual_dropdown.value
    department_name = department_dropdown.value
    assign_employee_to_department(department_name, employee_id)

# Event handler for the college assignment button click
def on_assign_to_college_clicked(b):
    employee_id = individual_dropdown.value
    college_name = college_dropdown.value
    assign_employee_to_college(college_name, employee_id)

# Event handler for the vendor assignment button click
def on_assign_to_vendor_clicked(b):
    employee_id = individual_dropdown.value
    vendor_name = vendor_dropdown.value
    assign_employee_to_vendor(vendor_name, employee_id)

# Attach the event handler to the buttons
assign_to_department_button.on_click(on_assign_to_department_clicked)
assign_to_college_button.on_click(on_assign_to_college_clicked)
assign_to_vendor_button.on_click(on_assign_to_vendor_clicked)

# Display the widgets and buttons
display(individual_dropdown)
display(department_dropdown)
display(assign_to_department_button)

display(college_dropdown)
display(assign_to_college_button)

display(vendor_dropdown)
display(assign_to_vendor_button)
```
